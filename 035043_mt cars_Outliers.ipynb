```python
# Highlighting outliers by different color 

def identify_outliers(x):
    q1, q3 = np.percentile(x, [30, 70])
    iqr = q3 - q1
    lower_limit = q1 - (1 * iqr)
    upper_limit = q3 + (1 * iqr) 
    l = []
    for r in x:
        if np.less(r, lower_limit):
             l.append('background-color: red')
        elif np.greater(r, upper_limit):
             l.append('background-color: green')
        else:
            l.append('background-color: blue')
    return l

df.style.apply(identify_outliers)
```

```python
# Highlighting outliers by box plot

import matplotlib.pyplot as plt

def identify_outliers(x):
    q1, q3 = np.percentile(x, [30, 70])
    iqr = q3 - q1
    lower_limit = q1 - (1 * iqr)
    upper_limit = q3 + (1 * iqr)
    outliers = []
    for r in x:
        if np.less(r, lower_limit) or np.greater(r, upper_limit):
            outliers.append(r)
    return outliers

for col in df.columns:
    outliers = identify_outliers(df[col])
    plt.boxplot(df[col], showfliers=False)
    plt.scatter(np.ones(len(outliers)), outliers, color='green')
    plt.title(col)
    plt.show()
```

```python
# Highlighting outliers by scatter plot

import matplotlib.pyplot as plt

def identify_outliers(x):
    q1, q3 = np.percentile(x, [30, 70])
    iqr = q3 - q1
    lower_limit = q1 - (1 * iqr)
    upper_limit = q3 + (1 * iqr)
    outliers_x = []
    outliers_y = []
    for i, r in enumerate(x):
        if np.less(r, lower_limit) or np.greater(r, upper_limit):
            outliers_x.append(i)
            outliers_y.append(r)
    return outliers_x, outliers_y

for col in df.columns:
    outliers_x, outliers_y = identify_outliers(df[col])
    plt.scatter(df.index, df[col], color='blue')
    plt.scatter(outliers_x, outliers_y, color='green')
    plt.title(col + ' Outliers')
    plt.show()
```
